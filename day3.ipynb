```python
# Ordered unique list
numbers = [5,3,5,2,3]

seen = set()
result = []

for num in numbers:
    if num not in seen:
        seen.add(num)
        result.append(num)

print(result)
```

```python
#Mini Assignment (Customer Orders Processing)
customers = [
    {"id":401, "name":"Anita", "city":"Pune", "orders":[120,80]},
    {"id":402, "name":"Bhavesh", "city":"Mumbai", "orders":[1500]},
    {"id":403, "name":"Celine", "city":"Pune", "orders":[50,40,60]}
]
```

```python
#1
high_value = []
cities = set()
total_sales = 0
follow_up = []
```

```python
#2
for customer in customers:

    # unique cities
    cities.add(customer["city"])

    # calculate customer's total
    customer_total = sum(customer["orders"])

    # add to company sales
    total_sales += customer_total

    # high-value orders
    for order in customer["orders"]:
        if order >= 1000:
            high_value.append(customer["id"])

    # follow-up customers
    if customer_total < 200:
        follow_up.append(
            (customer["id"], customer["name"])
        )


print("High-value customers:", high_value)
print("Unique cities:", cities)
print("Total sales:", format(total_sales, ".2f"))
print("Follow-up customers:", follow_up)
```
