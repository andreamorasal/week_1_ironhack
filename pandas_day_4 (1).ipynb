```python
#Load the dataset into a DataFrame called df. 

df = pd.read_csv("/Users/andreams/Downloads/nba.csv")
```

```python
# Display the first 5 rows. 

df.head(5)
```

```python
#Check the number of rows and columns. 

df.shape
```

```python
# List all column names. 
df.columns
```

```python
# Data Cleaning  Convert the Salary column to numeric (handle errors). 

df['Salary'] = pd.to_numeric(df['Salary']) 
```

```python
#Count how many missing (NaN) values are in each column. 

print(df.isnull().sum())
```

```python
#Drop rows where Salary is missing. 

df = df.dropna(subset=['Salary'])
```

```python
# Fill missing College values with "Unknown". 
df['College'] = df['College'].fillna('Unknown')
```

```python
# Filtering Data Show all players from Boston Celtics. 

boston_players = df[df["Team"] == "Boston Celtics"].sort_values(by="Name")
print(boston_players)
```

```python
# Show players whose Age is greater than 30. 

age_players = df[df["Age"] > 30]
print(age_players)
```

```python
#Filter players who play as PG (Point Guard). 
pg_players = df[df["Position"] == "PG"]
print(pg_players)
```

```python

#Find players with Salary greater than 5,000,000. 

high_salary = df[df["Salary"] > 5000000]
print(high_salary)
```

```python
# Sorting Sort players by Salary (highest first). 
salary_sort = df.sort_values(by="Salary", ascending=False)
print(salary_sort)
```

```python
#Sort players by Age (youngest first)

age_sort = df.sort_values(by="Age")
print(age_sort)
```

```python
#GroupBy Operations - Average salary per team
avg_salary_team = df.groupby("Team")["Salary"].mean()
print(avg_salary_team)
```

```python
#Count number of players in each team
player_count_team = df.groupby("Team").size()
print(player_count_team)
```

```python
#Maximum salary in each team
max_salary_team = df.groupby("Team")["Salary"].max()
print(max_salary_team)
```

```python
#Aggregations  Average age of players
print(df["Age"].mean())
```

```python
#Maximum salary in the dataset
print(df["Salary"].max())
```

```python
#Minimum weight
print(df["Weight"].min())
```

```python
#Column Operations Create Age_in_5_years column
df["Age_in_5_years"] = df["Age"] + 5
print(df)
```

```python
#Create Salary_in_Millions column
df["Salary_in_Millions"] = df["Salary"] / 1000000
print(df)
```

```python
#Team with highest total salary payout
highest_team_salary = df.groupby("Team")["Salary"].sum().idxmax()
print(highest_team_salary)
```

```python
#Position with highest average salary
highest_position_salary = df.groupby("Position")["Salary"].mean().idxmax()
print(highest_position_salary)
```
