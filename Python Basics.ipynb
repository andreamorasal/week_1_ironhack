```python
#Lab 12: Marks percentage calculator (optional preview: simple if/else)

m1 = "78"
m2 = "85"
m3 = "69"

m1 = int("78")
m2 = int("85")
m3 = int("69")

total = m1 + m2 + m3

percentage = (total/300)*100

print(f"Percentage:{percentage:.2f}%")

if percentage >= 35:
    print("Result: Pass") 
else: 
    print("Result: Fail")
```
